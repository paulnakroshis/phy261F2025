# Sun-Earth-Jupyter
## Background
In class, we have looked at a simulation (with the Euler-Cromer method) of the Earth and Moon. 
It turns out (those of you who have had Phy 331 will have seen this) that the two body system can be 
solved exactly. However, once you get to *three* bodies, there are very few analytical solutions; in general there is no analytical solution that covers all cases. In fact, the *n-body* problem is *the* problem in celestial mechanics, and we have to study it by simulation. 

In this project, I'm asking you to simulate the Sun-Earth-Jupyter system. This is question 4.16 in Giordano's book, so the handout I gave you is a good reference. Here is the statement of the question:

We also suggest that you give the Sun an initial velocity which makes the total momentum of the system exactly zero (so that the center of mass of the system remains fixed.)
