# Real Time Plotting with Makie
## An Example using GLMakie


```julia
using GLMakie

# Create observables for reactive updates (start empty to skip first point)
x = Observable(Float64[])
y = Observable(Float64[])

# Create the plot
fig = Figure()
ax = Axis(fig[1, 1], 
          limits=(0, 10, 0, 10),
          xlabel="X", 
          ylabel="Y",
          title="Real-Time with Makie")

# Plot the observables
scatter!(ax, x, y, markersize=20, color=:red)
lines!(ax, x, y, color=:blue, linewidth=2)  # Optional: add trail

display(fig)

# Update loop
for i in 1:200
    # Compute new position
    new_x = 5 + 3*cos(i/10)
    new_y = 5 + 3*sin(i/10)
    
    # Update observables correctly
    push!(x[], new_x)
    push!(y[], new_y)
    notify(x)  # Trigger the update
    notify(y)
    
    sleep(0.05)
end
```
